## Part 1: Explore the Data

Import the data and use Pandas to learn more about the dataset.

```python
import pandas as pd

df = pd.read_csv('Resources/client_dataset.csv')
df.head()
```

```python
# LAMBDA PRACTICE: LAMBDA USED WHEN YOU NEED TO RETURN A VALUE TO A SINGLE KNOWN VARIABLE
def multiply_args(x, y):
    return x * y
 
df = pd.DataFrame({"x": [1,2,3,4,5], "y":[3,3,3,3,3]})

# Lamda Syntax:
# variable_name = lambda arg1, arg1: return statement

# Lamda Funtion::
# multiply = lambda x, y: multiply_args

# Function Lambda call:
# multiply(4, 7)

# Apply Syntax: 
# variable = DataFrame.apply(method)

# Example:
# df1=df.apply(multiply_args)

# Lambda and Apply
# Wrong:
# df1["result"]=df.apply(lambda df["x"], df["y"]: multiply_args)
# Correct
df["result"]=df.apply(lambda row: multiply_args(row["x"], row["y"]), axis=1)
df
```

```python
# LAMBDA PRACTICE: LAMBDA USED WHEN YOU HAVE A CONDITION AND NEED TO RETURN A VALUE TO A SINGLE KNOWN VARIABLE
# # Create a column that calculates the subtotal for each line using the unit_price and the qty, 
def calculate_subtotal(row):
    return row['unit_price'] * row["qty"]

# df["line_subtotal"] = df.apply(calculate_subtotal, axis = 1 )
df["line_subtotal"] = df['unit_price'] * df["qty"]

#Add discount col : If line_subtotal > 1500 give 10% disc otherwise keep the same 
def calculate_discount(subtotal):
    if subtotal> 1500: 
        return subtotal* 0.10
    else: 
        return 0
    
# df["discount"]= df['line_subtotal'].apply(calculate_discount)

df["discount"]= df['line_subtotal'].apply(lambda total: total *0.10 if total> 1500 else total* 0)



subtotal_df = df[["client_id", "unit_price", "qty", "line_subtotal", "discount"]]
# subtotal_df = df.loc[df["client_id"] == 33615, ["client_id", "unit_price", "qty", "line_subtotal"]]

subtotal_df
```
